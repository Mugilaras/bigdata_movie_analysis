 I loaded the 6 dataset by giving a numerical value of the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets

dataset1 = pd.read_csv("/content/genome-scores.csv")
dataset2 = pd.read_csv("/content/genome-tags.csv")
dataset3 = pd.read_csv("/content/links.csv")
dataset4 = pd.read_csv("/content/movies.csv")
dataset5 = pd.read_csv("/content/ratings.csv")
dataset6 = pd.read_csv("/content/tags.csv")



```

this gives an understanding of how the dataset looks, so that we can get an idea what dataset we are working on

```python


# Print the head of each dataset

print("Dataset 1:")
print(dataset1.head())

print("\nDataset 2:")
print(dataset2.head())

print("\nDataset 3:")
print(dataset3.head())

print("\nDataset 4:")
print(dataset4.head())

print("\nDataset 5:")
print(dataset5.head())

print("\nDataset 6:")
print(dataset6.head())
```

it  provide a visual summary of the distribution of ratings in Dataset 5,
helping to understand the spread and concentration of different rating values.

```python
# Distribution of Ratings

plt.figure(figsize=(8, 6))
sns.histplot(dataset5['rating'], bins=5, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

```

his code snippet aims to provide a visual representation of the frequency distribution of different genres in Dataset 4.
The bar chart allows for an easy comparison of genre frequencies.

```python
# Bar chart for genre frequencies

plt.figure(figsize=(12, 6))
genre_counts = dataset4['genres'].str.split('|', expand=True).stack().value_counts()
genre_counts.plot(kind='bar')
plt.title('Genre Frequencies')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()

```

this code snippet aims to visually represent the most common words in movie titles from Dataset 4 using a word cloud,
offering an immediate sense of the prominent terms in the dataset.

```python
# Word cloud for movie titles

plt.figure(figsize=(12, 8))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(dataset4['title']))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Movie Titles')
plt.show()
```

provide a visual summary of the distribution of relevance scores in Dataset 1, helping to understand the spread and concentration of different relevance score values.


```python
# Distribution of Relevance Scores

plt.figure(figsize=(10, 6))
sns.histplot(dataset1['relevance'], bins=20, kde=True)
plt.title('Distribution of Relevance Scores')
plt.xlabel('Relevance Score')
plt.ylabel('Frequency')
plt.show()


```

this code snippet aims to provide a visual representation of the distribution of relevance scores for each tag in Dataset 1,
allowing for the comparison of relevance score variations across different tags.

```python
# Box plot to visualize distribution of relevance scores for each tag

plt.figure(figsize=(14, 8))
sns.boxplot(x='tagId', y='relevance', data=dataset1)
plt.title('Distribution of Relevance Scores for Each Tag')
plt.xlabel('Tag ID')
plt.ylabel('Relevance Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```
